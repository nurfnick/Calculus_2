```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

# Generating Torus Mesh
angle = np.linspace(0, 2 * np.pi, 32)
theta, phi = np.meshgrid(angle, angle)
r, R = 1., 2.
X = (R + r * np.cos(phi)) * np.cos(theta)
Y = (R + r * np.cos(phi)) * np.sin(theta)
Z = r * np.sin(phi)

# Visualizing the mesh
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.set_xlim3d(-2, 2)
ax.set_ylim3d(-2, 2)
ax.set_zlim3d(-2, 2)
ax.plot_surface(X, Y, Z, color = 'w', rstride = 1, cstride = 1)
plt.show()
```

# Task
Generate and plot the parametric equations of the graph y = (x-1)(x-2) rotated about the y axis.

## Derive parametric equations

### Subtask:
Determine the parametric equations for the surface of revolution by rotating the curve $y = (x-1)(x-2)$ about the y-axis.


**Reasoning**:
Determine the parametric equations for the surface of revolution by rotating the curve y = (x-1)(x-2) about the y-axis by expressing x and y in terms of a parameter u, and substituting into the standard parametric equations for a surface of revolution rotated about the y-axis.



```python
# The original curve is y = (x-1)(x-2)
# To parametrize the curve, let x = u.
# Then y = (u-1)(u-2) = u^2 - 3u + 2

# The standard parametric equations for a surface of revolution by rotating a curve
# x = f(u), y = g(u) about the y-axis are:
# X = f(u) * cos(theta)
# Y = g(u)
# Z = f(u) * sin(theta)

# In this case, f(u) = u and g(u) = u^2 - 3u + 2.
# So, the parametric equations for the surface are:
# X = u * cos(theta)
# Y = u^2 - 3u + 2
# Z = u * sin(theta)

# Print the parametric equations as strings
print("Parametric equations for the surface of revolution:")
print("X = u * cos(theta)")
print("Y = u^2 - 3u + 2")
print("Z = u * sin(theta)")
```

## Generate mesh data

### Subtask:
Use the derived parametric equations to generate the mesh data for the surface.


**Reasoning**:
Generate the mesh data for the surface using the derived parametric equations.



```python
# Create meshgrid for parameters u and theta
u = np.linspace(1, 2, 50)
theta = np.linspace(0, 2 * np.pi, 50)
u, theta = np.meshgrid(u, theta)

# Calculate X, Y, and Z coordinates using the parametric equations
X = u * np.cos(theta)
Y = u**2 - 3*u + 2
Z = u * np.sin(theta)

# Print the shapes to verify they are the same
print("Shape of X:", X.shape)
print("Shape of Y:", Y.shape)
print("Shape of Z:", Z.shape)
```

## Visualize the mesh

### Subtask:
Plot the generated mesh data to visualize the surface of revolution.


**Reasoning**:
Generate and plot the parametric equations of the graph y = (x-1)(x-2) rotated about the y axis.



```python
# Create a 3D plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the generated mesh data
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis', edgecolor='none')

# Add labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Surface of Revolution of y = (x-1)(x-2) about the Y-axis')

# Display the plot
plt.show()
```
