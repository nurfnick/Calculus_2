```python
def a(n):
  if n%3 == 0:
    return -1/(2*(n/3))
  else:
    return 1/(2*(2*(n//3)+n%3)-1)

sum([a(n) for n in range(1,100000)])
```

```python
val = 2
tolerance = 0.1
evenCounter = 0
def addSome(s,val,tolerance,evenCounter):
  while(abs(s-val)>tolerance):
    s+=b(2*evenCounter)
    evenCounter+=1
  return s,evenCounter
addSome(0,val,tolerance,evenCounter)

oddCounter = 0
def subtractSome(s,val,tolerance,oddCounter):
  s+=b(2*oddCounter+1)
  oddCounter+=1
  return s,oddCounter
subtractSome(2,val,tolerance,oddCounter)
```

```python
l = []
s = 0
evenCounter = 0
oddCounter = 0
for i in range(100):
  s,evenCounter = addSome(s,val,tolerance,evenCounter)
  l.append(evenCounter)
  s,oddCounter = subtractSome(s,val,tolerance,oddCounter)
  l.append(oddCounter)
print(l)
s
```
