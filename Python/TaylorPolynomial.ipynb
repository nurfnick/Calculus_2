```python
import math

def b(n,x): #each term in taylor expansion of e^x
  return 1/math.factorial(n)*x**n

def a(n,x): #the nth order polynomial for e^v
  return sum([b(i,x) for i in range(n+1)])
```

```python
a(100,2)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x_values = np.linspace(-10, 5, 400) # Range for x

# Calculate y values for e**x
y_exp = np.exp(x_values)

# Calculate y values for a(n,x) for n=1, 2, 25
y_a1 = [a(1, x) for x in x_values]
y_a2 = [a(2, x) for x in x_values]
y_a5 = [a(25, x) for x in x_values]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_exp, label='e^x', linestyle='--')
plt.plot(x_values, y_a1, label='a(1,x)')
plt.plot(x_values, y_a2, label='a(2,x)')
plt.plot(x_values, y_a5, label='a(25,x)')

plt.title('Comparison of e^x and a(n,x) functions')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.ylim(-1, 20)
plt.show()
```
