# Polar Coordinates

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate an array of angles (t) from 0 to 2*pi
t = np.linspace(0, 2 * np.pi, 1000)

# Calculate the radius (r) using the given equation
r = 1 + np.sin(t)

# Create a polar plot
plt.figure(figsize=(6, 6))
ax = plt.subplot(111, projection='polar')

# Plot the polar curve
ax.plot(t, r, color='skyblue', linewidth=2)

# Set title and display the plot
ax.set_title('Polar Plot of r = 1 + sin(t)')
plt.grid(True)
plt.show()
```

```python
r = np.cos(3*t)
t = np.where(r <= 0, t, t + np.pi)
r = np.abs(r)
# Create a polar plot
plt.figure(figsize=(6, 6))
ax = plt.subplot(111, projection='polar')

# Plot the polar curve
ax.plot(t, r, color='skyblue', linewidth=2)
```

```python
r = np.sqrt(np.sin(2*t))


# Create a polar plot
plt.figure(figsize=(6, 6))
ax = plt.subplot(111, projection='polar')

# Plot the polar curve
ax.plot(t, r, color='skyblue', linewidth=2)
```

```python
theta = np.linspace(0, 2 * np.pi, 1000)
r = 2*np.cos(theta)
theta = np.where(r >= 0, theta, theta + np.pi)
r = np.abs(r)
plt.polar(theta, r, color='skyblue', linewidth=2)
```
