## Lissajous Figures

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0, 100*np.pi, 10000)  # Create a list of evenly-spaced numbers over the range

x = np.sin(3*t)
y = np.cos(7*t)

plt.plot(x, y)
plt.show()
```

## Cycloid

```python
t = np.linspace(0, 8*np.pi, 1000)  # Create a list of evenly-spaced numbers over the range
r = 1

x = r*t -r*np.sin(t)
y = r-r*np.cos(t)

plt.plot(x, y)
plt.show()
```

## First Calculus Example

```python


t = np.linspace(-2, 2, 1000)  # Create a list of evenly-spaced numbers over the range

x = t**2
y = t**3 -3*t

plt.plot(x, y)
plt.show()
```
