# Parametric Curves

```python
from sympy import *
x,t,n = symbols(('x','t','n'))
```

```python
plot_parametric(t**4-t**3+7,t,(t,-5,5))
```

```python
entry = (-1)**n*x**(2*n+5)/(factorial(n)*factorial(n+5)*2**(2*n+5))
entry
```

```python
partsum1 = entry.subs(n,0)+entry.subs(n,1)
partsum1
```

```python
partsum10 = 0

for i in range(50):
  partsum10 += entry.subs(n,i)
```

```python
plot(besselj(5,x),(x,0,10))
```

```python
plot(partsum10,(x,0,10))
```
